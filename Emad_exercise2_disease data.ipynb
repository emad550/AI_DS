```
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```
#import the dataset from the source
dlink = "https://raw.githubusercontent.com/caravanuden/cardio/master/cardio_train.csv"
```

```
#create data frame for the date
data= pd.read_csv(dlink,sep=';')
```

```
#read the first 5 rows
data.head()
```

```
#we can check data types and more information about our data
data.info()
```

```
#check null values for our data 
data.isnull().sum()
```

```
# creating the trsnspose of the description of the Dataframe and then showing it
data.describe().T
```

```
# we can check the distribution of our data
data.hist(figsize=(20,20))
```

```
#conver age to years 
data.age=(data.age/365).astype('int')
```

```
#we can go in depth by checking  how impact of our features
#lets start with age by cardio
# we will discover that we have realtion between elder age and cardio 
sns.countplot(data.age,hue=data.cardio)
```

```
#checking our categorical data 
data_categorical = data.loc[:,['cholesterol','gluc', 'smoke', 'alco', 'active']]
sns.countplot(x="variable", hue="value",data= pd.melt(data_categorical));
```

```
#check age versus weight for man and women 
sns.barplot(data.age, data.weight,hue=data.gender)
```

```
#checking the gender and weight and how it correlate with cardio
sns.barplot(data.gender, data.weight,hue=data.cardio)
```

```
#how the smoking of gender be distributed 
sns.barplot(data.gender,data.smoke,hue=data.cardio)
```

```
#gender versus cardio
sns.countplot(data.gender,hue=data.cardio)
```

```
#how distibution of cardio 
data['cardio'].value_counts().plot(kind='bar')
```

```
# we can notice her that alco may be will not have efect or cardio
sns.countplot(data.alco,hue=data['cardio'])
```

```
# we will notice her that when cholesterol high will be high effect of cardio 
sns.countplot(data.cholesterol,hue=data['cardio'])
```

```
#correlation between features with cardio 
# we notice her we have some positive and negative correlation
plt.figure(figsize=(12,10))
sns.heatmap(data.corr(), annot=True,cmap ='RdYlGn') 
```
